```python
# (c)

# polting normal distribution curve and generating data

import matplotlib.pyplot as plt
import scipy.stats
import numpy as np


x_min = 0.0
x_max = 16.0

mean =-1 
std =2 
x = np.linspace(x_min, x_max, 100)
y = scipy.stats.norm.pdf(x,mean,std)
plt.plot(x,y, color='coral')
plt.grid()
plt.xlim(x_min,x_max)
plt.ylim(0,0.25)
plt.title('normal distribution curve ',fontsize=10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Question 4/ program

```python
# (a)

import networkx as nx
import pandas as pd
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


df = pd.DataFrame({'from':['v1','v2','v3'],'to':['v2', 'v3', 'v4']})
 
G1= nx.from_pandas_edgelist(df,'from','to',create_using=nx.Graph())
 
nx.draw(G1,with_labels=True,node_size=1500,alpha=0.3,arrows=True)
plt.show()
```

```python
# (b)
import networkx as nx
import pandas as pd
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


G2= nx.from_pandas_edgelist(df,'from','to',create_using=nx.DiGraph() )
 
nx.draw(G2, with_labels=True,node_size=1500,alpha=0.3,arrows=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x=[1,2,3,4,5,6]
y=[1,2,7,3,5,9]
y2=[2,4,6,8,3,5]


fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title('Plot Graph',fontsize=20)


ax.set_xlabel('.........x..........',fontsize=20)
ax.set_ylabel('.........y..........',fontsize=20)

plt.plot(x,y,label='curve1')
plt.plot(x,y2,label='curve2')

ax.legend()
```

```python
# (a)

# draw undirected graph

import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3,4,5])

G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(4,5)])

nx.draw(G)
```
