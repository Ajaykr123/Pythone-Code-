```python
# : (a) Circle 1: x2 + y2 = 3


r1 = np.sqrt(3)
x1 = np.linspace(-r1,r1,10000)
y1 = np.sqrt(-x1**2+r1**2)
plt.plot(x1, y1, 'b')
plt.plot(x1, -y1, 'b')
plt.gca().set_aspect('equal')
plt.show()
```

```python
# (b)  Circle 2: x2 + y2 = 2

r2 = np.sqrt(2)
x2 = np.linspace(-r2,r2,1000)
y2 = np.sqrt(-x2**2+r2**2)
plt.plot(x2, y2, 'g')
plt.plot(x2, -y2, 'g')
plt.gca().set_aspect('equal')
plt.show()
```

```python
#  (c) Red colour between Circle 1 and Circle 2 ,Green colour inside Circle 1.


circle1 = plt.Circle((0, 0),radius = np.sqrt(3), color='red')
circle2 = plt.Circle((0, 0),radius = np.sqrt(2), color='green')

plt.gca().add_patch(circle1)
plt.gca().add_patch(circle2)
plt.gca().set_aspect(1)

plt.grid()
plt.plot()   
plt.show()
```

7. Draw a figure that consists of the following subfigures: (a) Line 1: 6x âˆ’ 9y + 2 = 0
(b) Line 2: 6xâˆ’9y+2 = 1 (c) Line 3: 6xâˆ’9y+2 = âˆ’1 (d) A segment that connects
6x âˆ’ 9y + 2 = 1 and 6x âˆ’ 9y + 2 = âˆ’1. (e) Blue colour above 6x âˆ’ 9y + 2 = 1
and yellow colour below 6x âˆ’ 9y + 2 = âˆ’1.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(1,200,200)

plt.plot(x,(6*x+2)/9,linestyle='-')

plt.plot(x,(6*x+1)/9,linestyle='-')

plt.plot(x,(6*x+3)/9,linestyle='-')
```

```python
x= np.linspace (-2, 2, 200)

y1= (6*x + 2)/9
y2= (6*x + 1)/9
y3= (6*x + 3)/9

fig,ax= plt.subplots()
plt.xlabel('x-axis')
plt.ylabel('y-axis')
ax.plot(x, y1, label= 'line 1', color= 'red')
ax.plot(x, y2, label= 'line 2', color= 'black')
ax.plot(x, y3, label= 'line 3', color= 'blue')

# line segment between line 2 and line 3:
a=0
   # point in line 2:
b= (6*a + 1)/9   
   # point in line 3:
c= (6*a + 3)/9 
print("points of line 1:\t" + str(a) + "," + str(b))
print("points of line 2:\t" + str(a) + "," + str(c))
point1 = [a, a]
point2 = [b, c]

plt.scatter(point1, point2, color='black')
plt.plot(point1, point2, color= 'black')

ax.fill_between(x, y1, y2, facecolor='green', interpolate=True)
ax.fill_between(x, y1, y3, facecolor='yellow', interpolate=True)
plt.legend()
plt.show()
```
