```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_gaussian_quantiles
```

```python
X1, y1 = make_gaussian_quantiles(mean=(2.,4.,6.), cov=(1.,1.,1.), n_samples=1000, n_features=3, n_classes=2, random_state=1)
X1 = pd.DataFrame(X1,columns=['A1','A2','A3'])
y1 = pd.Series(y1)
X1.head()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(20,6))
sns.distplot(X1[y1==1]['A1'],bins=50,label="True Class",ax=ax1)
sns.distplot(X1[y1==0]['A1'],bins=50,label="False Class",ax=ax1)
ax1.set_xlabel("1st Dim")
ax1.legend()

sns.distplot(X1[y1==1]['A2'],bins=100,label="True Class",ax=ax2)
sns.distplot(X1[y1==0]['A2'],bins=100,label="False Class",ax=ax2)
ax2.set_xlabel("2nd Dim")
ax2.legend()
plt.show()
```
